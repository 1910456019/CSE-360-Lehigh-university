```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Example data: Replace these with your actual data
u = np.array([76, 94, 105, 115, 123, 122, 118, 122, 135, 138])
one_over_Sb = np.array([0.0001118, 0.000178, 0.000217, 0.000286, 0.000333,
                        0.000357, 0.000435, 0.000476, 0.000513, 0.000541])
d = np.array([25.612, 28.625, 30.675, 32.536, 34.597, 36.663, 39.324, 41.497, 44.463, 48.524])

# Stack the features into a single matrix for scikit-learn
X = np.column_stack((u, one_over_Sb))

# Create the regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, d)

# The coefficients for u and 1/Sb
a = model.coef_[0]
b = model.coef_[1]

# The intercept (c)
c = model.intercept_

print(f'The parameter a is approximately: {a}')
print(f'The parameter b is approximately: {b}')
print(f'The parameter c is approximately: {c}')
```

```python
# wraped helper function for using the parameter I caculated to predict distance "d" by
# given input "coordinate - center pixel" and "pixel size"

def calculate_distance(cx, pixels):
    # Assuming the area is a proxy for S_b and inverting it as 1/S_b
    S_b = pixels
    one_over_Sb = 1.0 / S_b

    # Use the regression equation to calculate d
    d = a * cx + b * one_over_Sb + c
    return d


# ------------------------------------------
# Test 1 --- data from measurement from excel data
# ------------------------------------------

# measure data
pixelSize = 8460
coord_u = 76
# Calculate the distance d
distance = calculate_distance(coord_u, pixelSize)
# give output
print("(Unit in cm) Real Measruement: 25.612, predict:",distance)


# ------------------------------------------
# Test 2 --- data from measurement from excel data
# ------------------------------------------

# measure data
pixelSize2 = 1850
coord_u2 = 138
# Calculate the distance d
distance2 = calculate_distance(coord_u2, pixelSize2)
# give output
print("(Unit in cm) Real Measruement: 48.524, predict:",distance2)

```
