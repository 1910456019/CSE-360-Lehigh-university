# Building a Robot Simulator in 2D

```python
# Perfect sensor
def sense(x):
    return x
```

### Euler integration

```python
def simulate(Δt, x, u):
    x += Δt * u
    return x
```

### Control policy

```python
def control(t, y):
    ### WRITE YOUR CONTROL POLICY HERE:
    kx = 1
    ky = 1

    px = 0;
    py = 0;

    if t < 2:
      px = 3
      py = 3
    elif (t >= 2) and (t < 4):
      px = 3
      py = -3
    elif (t >= 4) and (t < 8):
      px = -3
      py = 3
    elif (t >= 8) and (t < 10):
      px = -3
      py = -3

    ux = kx * (px - y[0])
    uy = ky * (py - y[1])

    # print(px,py,t)

    return array([ux, uy])
```

### Running the simulator

```python
# final time to be a multiple of 2*pi for 8 curve
tf = 10
Δt = 0.1    # Time step
time = linspace(0.,tf, int(tf / Δt) + 1)  # Time interval


# initial condition
x = array([0., 0.])


x_log = [copy(x)]

for t in time:
    y = sense(x)
    u = control(t, y)
    x = simulate(Δt, x, u)
    x_log.append(copy(x))

x_log = array(x_log)
```
